### Dataset

A publication on the American Association for the Study of Liver Diseases AASLD, “the leading organization of scientists and health care professionals committed to preventing and curing liver disease” (AASLDpubs), presented the findings of a 20 year study on non-alcoholic fatty liver disease NAFLD. NAFLD occurs when there is build up of fats in the liver in people that drink little to no alcohol.  
The study was done in a  “community cohort of all adults diagnosed with NAFLD in Olmsted County, Minnesota, between 1997 and 2014 using the Rochester Epidemiology Project database” (Allen et al.). All medical data collected from the medical providers in this community was linked through an ID number in a system.  
The dataset from this study is named Non-alcohol fatty liver disease (NAFLD) found in Kaggel.com (Singh). The data file contains the baseline information about the characteristics of the subject.  
The following are the variables of the dataset:  
1. **Subject identifier (Unnamed):**  an arbitrary number given to each subject across medical services. It doesn’t add any value to the predictions.  
2. **ID number for each sample (id):** identifier integer for each subject.   
3. **Age (age):** the age of the subject at the time of the entry to the study. This can influence the metabolic process of the body.  
4. **Male (male):** whether the subject is a male. If the person is a male it is represented with 1(true) otherwise with 0 (false).  
5. **Weight (weight):** the weight of the person is represented in kilograms. It is shown that the weight of the person can have big effects on the overall health of the people.  
6. **Height (height):** the height of the person represented in centimeters. This affects the BMI calculation in combination with the weight.  
7. **BMI (bmi):** the Body Mass Index is a measurement of fat calculated using weight and height. It is a ratio that better describes the amount of fat in a person's body.  
8. **Case.ID (case.id):** is a number id  that represents the NAFLD cas. However, this column does not add value to the predictions.  
9. **Futime (futime):** it is the time to death or to the last medical follow up.  
10. **Status (status):** This column represents the status of the person at the last follow up. It is represented with 0 as alive and 1 as dead.

### Statistical Analysis: Central Tendency and Variability
[Table of Contents Link](#Table-of-Contents)

#### Central Tendency

Central tendency is a way to describe a value that represents a data set.  
There are three methods that describe the measures of central tendency:
1. Mean 
2. Median
3. Mode   

In the following section I will calculate the measures of central tendency for the data set. 

```python
# function for calculating the central tendency measurments
def CalculateCT(cols, func, ct_name, dictny):
    print(ct_name.upper())
    col_arr = []
    for col in cols:
        print('  ', col.upper(), ': ', func(nafld_median[col]))
        col_arr.append(func(nafld_median[col]))
    if ct_name == 'mean':
        dictny['mean'] = col_arr
    elif ct_name == 'median':
        dictny['median'] = col_arr
    else:
        dictny['mode'] = col_arr
```

```python
# Calculate the central tendency of the data
stats_cols = ['age', 'weight', 'height', 'bmi']
ct_dictny = {'mean': None, 'median': None, 'mode': None}
# Calculate the mean
print(CalculateCT(stats_cols, statistics.mean, 'mean', ct_dictny))
# Calculate the median
print(CalculateCT(stats_cols, statistics.median, 'median', ct_dictny))
# Calculate the mode
print(CalculateCT(stats_cols, statistics.mode, 'mode', ct_dictny))
```

#### Variability

Variability is how the data is spread across the ranges.  
There are three methods for describing the dataset variability:
1. Range 
2. Variance
3. Standard deviation

```python
# Function to calculate the range of a column 
def CalcRange(col):
    return max(col) - min(col)
```

```python
# function for calculating the variability measurments
def CalculateVar(cols, func, var_name, dictny):
    print(var_name.upper())
    col_arr = []
    for col in cols:
        print('  ', col.upper(), ': ', func(nafld_median[col]))
        col_arr.append(func(nafld_median[col]))
    if var_name == 'range':
        dictny['range'] = col_arr
    elif var_name == 'variance':
        dictny['variance'] = col_arr
    else:
        dictny['sd'] = col_arr
```

```python
# Calculate the variability of the data
var_dictny = {'range': None, 'variance': None, 'sd': None}
# Calculate the range
print(CalculateVar(stats_cols, CalcRange, 'range', var_dictny))
# Calculate the variance
print(CalculateVar(stats_cols, statistics.variance, 'variance', var_dictny))
# Calculate the standard deviation
print(CalculateVar(stats_cols, statistics.stdev, 'standard deviation', var_dictny))
```

```python
# Function to format the subplots
def FormatPlots(ax1, ax2, xs, ys, q1, q3, inter_qr, median, pctg):
    #Colors the quartiles on the probability density function graph
    ylims = ax1.get_ylim()
    ax2.fill_between(xs, 0, ys, where=(xs >= q1 - 1.5 * inter_qr) & (xs <= q3 + 1.5 * inter_qr), color='white', alpha=0.3)
    ax2.fill_between(xs, 0, ys, where=(xs >= q1) & (xs <= q3), color='white', alpha=0.5)
    
    # Quartiles
    ax1.text(q1 - 2 * inter_qr, -0.8, "%.2f" % pctg['L.outlier'] + '%', ha='center', va='center', color='#FA8900', size=7)
    ax1.text(q1 - (inter_qr / 2), -0.8, "%.2f" % pctg['25'] + '%', ha='center', va='center', color='#FA8900', size=7)
    ax1.text(q1 + (inter_qr / 2), -0.8, "%.2f" % pctg['IQR'] + '%', ha='center', va='center', color='#FA8900', size=7)
    ax1.text(q3 + (1.5 * inter_qr / 2), -0.8, "%.2f" % pctg['75'] + '%', ha='center', va='center', color='#FA8900', size=7)
    ax1.text(q3 + 2 * inter_qr, -0.8, "%.2f" % pctg['R.outlier'] + '%', ha='center', va='center', color='#FA8900', size=7)
    
    ax1.text(q1, 0.6, 'Q1', ha='center', va='center', color='black')
    ax1.text(median, 1.0, 'IQR', ha='center', va='center', color='black')
    ax1.text(q3, 0.6, 'Q3', ha='center', va='center', color='black')
    ax1.text(median, -0.6, 'Median', ha='center', va='center', color='black')
    
    # Whiskers
    ax1.text(median - 1.5*inter_qr, 0.4, 'Q1 - 1.5*IQR', ha='center', va='center', color='black')
    ax1.text(q3 + 1.5*inter_qr, 0.4, 'Q3 + 1.5*IQR', ha='center', va='center', color='black')
    ax1.set_ylim(-1, 1.5)
    ax1.set(xlabel=None)
```

```python
# Function to creates the plots
def CreatesPlots(feat, col, xys):
    sns.kdeplot(nafld_median[feat], ax=ax2, color='black')
    kde_line = ax2.lines[0]
    xys['xs'] = kde_line.get_xdata()
    xys['ys'] = kde_line.get_ydata()

    #............. I wrote this code
    # Creates the box plot with the 25, 50, 75 quartiles 
    sns.boxplot(x=nafld_median[feat], color=col, whis=1.5, ax=ax1)

    # Creates a histogram of the age
    ax2.hist(nafld_median[feat], bins=20, color=col, edgecolor='#404040', density=True)
    #............. end of code I wrote
```

```python
# function to add labels to the histogram
def HistLabels(feat):
    plt.xlabel(feat.capitalize())
    plt.ylabel('Density')
    plt.axvline(stats_ct_DF.loc[feat].loc['mean'], linestyle=':', c='r', label='mean')
    plt.axvline(stats_ct_DF.loc[feat].loc['median'], linestyle=':', c='yellow', label='median')
    plt.axvline(stats_ct_DF.loc[feat].loc['mode'], linestyle=':', c='black', label='mode')
    plt.legend(loc='best', facecolor='lightgray')
```

```python
def CalcPercentage(data_name, q1, q3, iqt):
    pge_qr = {'L.outlier': None, '25': None, 'IQR': None, '75': None, 'R.outlier': None}
    d_ser = nafld_median[data_name]
    
    ser_len = len(d_ser)
    l_outl = len(d_ser.loc[lambda cell : (cell < q1 - 1.5 * iqt)])
    q1_len = len(d_ser.loc[lambda cell : (cell >= q1 - 1.5 * iqt) & (cell < q1)])
    mid_len = len(d_ser.loc[lambda cell : (cell >= q1) & (cell < q3)])
    q3_len = len(d_ser.loc[lambda cell : (cell >= q3) & (cell < q3 + 1.5 * iqt)])
    r_outl = len(d_ser.loc[lambda cell : (cell >= q3 + 1.5 * iqt)])
    
    pge_qr['L.outlier'] = l_outl / len(d_ser) * 100 
    pge_qr['25'] = q1_len / len(d_ser) * 100 
    pge_qr['IQR'] = mid_len / len(d_ser) * 100 
    pge_qr['75'] = q3_len / len(d_ser) * 100 
    pge_qr['R.outlier'] = r_outl / len(d_ser) * 100 

    return pge_qr
```

```python
# Creates subplots for the age data distribution
fig, (ax1, ax2) = plt.subplots(nrows=2, sharex=True)
fig.suptitle('Histogram of the Age')
fig.set_figwidth(10)

# calculate quartiles 
q1, median, q3 = np.percentile(nafld_median['age'], [25, 50, 75]) 
inter_qr = q3 - q1                            
pge_qr = CalcPercentage('age', q1, q3, inter_qr)

#calls the function to create the plots
xys = {'xs':None, 'xy':None}
CreatesPlots('age', '#761457', xys)

# Call function to format the histogram
HistLabels('age')

# Call function to format the plots
FormatPlots(ax1, ax2, xys['xs'], xys['ys'], q1, q3, inter_qr, median, pge_qr)

plt.show()
```

### Feature Engineering Second Itaration
[Table of Contents Link](#Table-of-Contents)

During the analysis comparisons we can see that the three main variables have too many outliers that can tilt the results of the prediction. Specially the BMI, its outliers are located too far from the mean. Therefore, it is important to find ways to minimize the impact on the analysis.

First, we can understand if there are values that may not be consider real in each feature

```python
means = [nafld_median[i].mean() for i in nafld_median.columns]
stds = [nafld_median[i].std() for i in nafld_median.columns]
std_l_lim = [i - j * 3 for i, j in zip(means, stds)]
std_r_lim = [i + j * 3 for i, j in zip(means, stds)]

iqrs = [nafld_median[i].quantile(.75) - nafld_median[i].quantile(.25) for i in nafld_median.columns]
iqr_space = [i * 1.5 for i in iqrs]
iqr_l_lim = [nafld_median[i].quantile(.25) - j for i, j in zip(nafld_median.columns, iqrs)]
iqr_r_lim = [nafld_median[i].quantile(.75) + j for i, j in zip(nafld_median.columns, iqrs)]

mins = [nafld_median[i].min() for i in nafld_median.columns]
maxs = [nafld_median[i].max() for i in nafld_median.columns]

stats_table = pd.DataFrame(zip(means, stds, std_l_lim, std_r_lim, iqrs, iqr_space, iqr_l_lim, iqr_r_lim, mins, maxs), 
             index=nafld_median.columns, 
             columns=['mean', 'std', 'mean-std(3)', 'mean+std(3)', 'iqr', 'iqr(1.5)', 'Q1 - iqr(1.5)', 'Q3 + iqr(1.5)', 'min', 'max'])
stats_table
```

> From the previous chart we can see that the values for the main features are not extreme or impossible to occur in real life, however they are too few cases to accurately represent the group at that level. 

In the following two sections we are going to analyze what case is better suited to remove the cases with outliers. As an experiment, the rows are removed using the following two cases: 
1. The quartiles and the interquartile multiplied by 1.5
2. The three standard deviations 

* **Quartiles Method**

```python
# Function to remove rows using the quartiles method
def RemoveRowsByQR(dta, stat_tab, feat):
    new_tab = []
    for i, val in enumerate(dta[feat]):
        if (val > stat_tab['Q1 - iqr(1.5)'][feat] and val < stat_tab['Q3 + iqr(1.5)'][feat]):
            new_tab.append(dta.iloc[i])
    new_tab = pd.DataFrame(new_tab)
    return new_tab
```

```python
print('Quartiles Method:')
print('  The length of the original table is: ', len(nafld_median))
print('  The length of the new table is: ', len(RemoveRowsByQR(nafld_median, stats_table, 'weight')))
```

* **Standard Deviation Method**

```python
# Function to remove rows using the standard deviation method
def RemoveRowsBySTD(dta, stat_tab, feat):
    new_tab = []
    for i, val in enumerate(dta[feat]):
        if (val > stat_tab['mean-std(3)'][feat] and val < stat_tab['mean+std(3)'][feat]):
            new_tab.append(dta.iloc[i])
    new_tab = pd.DataFrame(new_tab)
    return new_tab
```

```python
print('Standar Deviation Method:')
print('  The length of the original table is: ', len(nafld_median))
print('  The length of the new table is: ', len(RemoveRowsBySTD(nafld_median, stats_table, 'weight')))
```

> From this experiment, we can see that the rows removed by using the standard deviation are fewer cases than the ones removed by the quartiles method, therefore it is more appropiated to remove only the most extrems cases.

* **Removing the rows by Standard Deviation**

```python
# Remove rows with extrems values in the weight column
print('Remove from Weight Feature: ')
nafld_clean = RemoveRowsBySTD(nafld_median, stats_table, 'weight')
print('  The length of the original table is: ', len(nafld_median))
print('  The length of the new table is: ', len(nafld_clean))

# Remove rows with extrems values in the height column
print('Remove from Weight Feature: ')
nafld_clean = RemoveRowsBySTD(nafld_clean, stats_table, 'height')
print('  The length of the original table is: ', len(nafld_clean))
print('  The length of the new table is: ', len(nafld_clean))

# Remove rows with extrems values in the bmi column
print('Remove from Weight Feature: ')
nafld_clean = RemoveRowsBySTD(nafld_clean, stats_table, 'bmi')
print('  The length of the original table is: ', len(nafld_clean))
print('  The length of the new table is: ', len(nafld_clean))
print( )
print('The TOTAL of rows removed with extrems outliers are: ', len(nafld_median) - len(nafld_clean))
print('The percentage of remaining data is: ', '%.3f' % (len(nafld_clean) / len(nafld_median) * 100), '%')
```

The confusion matrix summarizes the number of correct and incorrect predictions made by the classifier. It provides insight into the model's ability to correctly identify both surviving and deceased patients.
Confusion matrix to visualize the performance a classification algorithm by showing:
1. True positives
2. False Positives
3. True Negatives
4. False Negatives

```python
from sklearn.metrics import (
    confusion_matrix,
    ConfusionMatrixDisplay,
    classification_report,
    precision_score,
    recall_score,
    f1_score,
    balanced_accuracy_score,
    roc_auc_score
)

conf = confusion_matrix(y_test_log, y_pred)

TN, FP, FN, TP = conf.ravel()

print(f"True Positive (TP): {TP}")
print(f"False Positive (FP): {FP}")
print(f"True Negative (TN): {TN}")
print(f"False Negative (FN): {FN}")

disp = ConfusionMatrixDisplay(confusion_matrix=conf)
disp.plot(cmap="Blues")
plt.title("Confusion Matrix")
plt.show()
```
